The below assignment produces the data on Top 10 Entrepreneurs in the World.
The data of the entrepreneurs has been srapped from 3 different sources as expected.
1. Web scrapping
2. Web API(Twitter API in below case)
3. CSV file(was collected from a databse made)

The sources for the website and csv file is as below. The explanation of the web scrapping follows.

Web URL - https://www.ranker.com/list/most-successful-entrepreneurs-in-the-world/business-and-company-info?page=1

CSV file - Created from all the Wiki pages of the respective entrepreneurs

Twitter API

```python
from bs4 import BeautifulSoup as bs
from requests import get
import pandas as pd
import tweepy
import csv
import pandas as pd
import re
```

```python
num=[1,2,3,4,5,6]
url="https://www.ranker.com/list/most-successful-entrepreneurs-in-the-world/business-and-company-info?page="
ent_name=[]
ent_age=[]
ent_desc=[]
ent_birthplace=[]
ent_profession=[]

for i in num:
    url1=url+str(i)
    print(url1)
#     webscrapper(url1, ent_name, ent_age, ent_desc, ent_birthplace, ent_profession)
    response = get(url1)
    html_soup = bs(response.text,'html.parser')## python's in built library HTML parser
    id_check = html_soup.find("article", class_="list__article list__section layout--adjust relative whiteBg list--headerEmpty list--noVotes")
    ent_container = id_check.find_all("div", class_="listItem__content--blog relative")
    len(ent_container)
    container = ent_container[0]
    sub_container = container.find_all("p", class_ = "listItem__property relative")
    age = sub_container[0].text[6:]
#     ent_name=[]
#     ent_age=[]
#     ent_desc=[]
#     ent_birthplace=[]
#     ent_profession=[]
    
    for container in ent_container:
        
        # Name of the Entrepreneurs
        try:
            name = container.h2.a.text
            ent_name.append(name)
        except:
            break
        
        #Description of the ent
        desc=container.find("p", class_="listItem__content--wiki listItem__content--wiki-hasBlather").text
        ent_desc.append(desc)
        
        sub_container = container.find_all("p", class_ = "listItem__property relative")
        
        ent_age.append(int(sub_container[0].text[6:]))
        ent_birthplace.append(sub_container[1].text[13:])
        try:
            ent_profession.append(sub_container[2].text[13:])
        except:
            ent_profession.append(" ")
    
#     return ent_name
#     return ent_age
#     return ent_desc
#     return ent_birthplace
#     return ent_profession
print(ent_name)
print(ent_age)
print(ent_desc)
print(ent_birthplace)
print(ent_profession)
```

```python
ent_df = pd.DataFrame({"Names":ent_name,
                        "Description":ent_desc,
                        "Age":ent_age,
                        "Birthplace":ent_birthplace,
                        "Profession":ent_profession
                        })
ent_df
```
